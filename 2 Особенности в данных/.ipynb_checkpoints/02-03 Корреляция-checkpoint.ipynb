# Корреляция

```python
import numpy as np
import pandas as pd
from scipy import stats

import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("ggplot")
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
# Новая переменная - процент чаевых от счета
df["tip_pct"] = 100*df["tip"]/df["total_bill"]
```

Посмотрим как между собой взаимосвязаны счет, чаевые и пол клиента

```python
sns.pairplot(df, hue="sex", height=2.5, kind="scatter");
```

- По графикам видно, что общий счет и величина чаевых зависят друг от друга, чем больше счет, тем больше чаевых оставляют, с долей подобной взаимосвязи не видно
- Пол никак не влияет на то, какое количество чаевых оставлено, красные и синие точки на графиках распределены равномерно, плотности на диагонали выгглядят похожим образом

#### Посмотрим разбиение по дням недели

```python
sns.catplot("day", "total_bill", "sex", data=df, kind="box",size=6);
```

- Почему-то в пятницу мужчины платят больше чаевых, чем женщины. Возможно, что это как-то связано с пятничными свиданиями и с древней традицией-счет оплачивает мужчина. Верна ли эта гипотеза? 

#### Посмотрим сколько наблюдений приходится на какой день недели.

```python
sns.catplot("day", "total_bill","sex", data = df, size=6);
```

В итоге получаем, что наши предыдущие выводы оказались ложными. Слишком мало данных в пятницу, чтобы делать такие выводы

#### Можно посмотреть на взаимосвязь только двух переменных

```python
df_log = df[["total_bill", "tip_pct"]].apply(np.log)
sns.jointplot("total_bill", "tip_pct", data=df_log, kind="reg");
```

#### Посмотрим на ковариационную матрицу

```python
df.corr(method="pearson") # Корреляция Пирсона
```

```python
df.corr(method="spearman") # Корреляяция Спирмана
```

Можно визуализировать матрицу с помощью тепловой карты

```python
sns.set(font_scale=1.3)
plt.subplots(figsize=(7,7))
sns.heatmap(df.corr(method="spearman"), square=True,
           annot=True, fmt=".2f", linewidths=0.1, cmap="RdBu");
```

#### Пример с рангами из презентации

Корреляция Спирмэна - это корреляция Пирсона, которую посчитали между рангами

```python
x = np.array([10,8,6,7,4,10,9,5])
y = np.array([9,9,4,5,6,8,10,7])
stats.spearmanr(x,y).correlation
```

```python
rx = np.array([7.5,5,3,4,1,7.5,6,2])
ry = np.array([6.5,6.5,1,2,3,5,8,4])
stats.pearsonr(rx,ry)[0]
```
