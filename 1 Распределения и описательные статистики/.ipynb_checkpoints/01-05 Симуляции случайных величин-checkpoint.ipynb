# Симуляции случайных величин 

```python
import numpy as np
import pandas as pd
from scipy import stats # Модуль для статистики

import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("ggplot")
```

## Генерация случайных величин

Будем использовать библиотеку scipy и модуль статистики. Сгенерируем нормальное распределение. $loc$ - мат. ожидание, $scale$ - среднеквадратичное отклонение.

```python
norm_rv = stats.norm(loc=0, scale=1) # Задали генератор
```

Различные методы для генератора:
- rvs - Сгенерирует нам выборку из распределения объёма size
- cdf - Вычислит значечние функции распределения в указанной точке (cumulative distribution function)
- pdf - Вычислит значение плотности распределения в указанной точке (probability density function)
- ppf - Вычислит квантиль, указанного уровня

```python
sample = norm_rv.rvs(1000) # генерация 1000 значений
```

## Плотность распределения и функция распределения

```python
norm_rv.pdf(1) # Узнали f(1)
```

Построим карнтину для плотности расрпделения, f(x)

```python
x = np.linspace(-3, 3, 100) # array диапазон
pdf = norm_rv.pdf(x)

plt.plot(x, pdf)
plt.ylabel("$f(x)$")
plt.xlabel("$x$")

# Нарисуем точки f(1), f(2)
plt.scatter([1, 2], [norm_rv.pdf(1), norm_rv.pdf(2)], color="blue")
```

Теперь узнаем значение функции распределения в точке 1 (вероятность, что $X<=1$):

```python
norm_rv.cdf(1)
```

```python
x = np.linspace(-3, 3, 100) # array диапазон
pdf = norm_rv.pdf(x)

plt.plot(x, pdf)
plt.ylabel("$f(x)$")
plt.xlabel("$x$")

# Нарисуем точки f(1), f(2)
plt.scatter([1, 2], [norm_rv.pdf(1), norm_rv.pdf(2)], color="blue")

# Добавим штриховку
xq = np.linspace(-3, 1)
yq = norm_rv.pdf(xq)
plt.fill_between(xq, 0, yq, color="blue", alpha=0.2)
# Синия линия 
plt.axvline(1, color="blue", linestyle="--", lw=2)
```

Теперь построим график для функции распределения.

```python
x = np.linspace(-3,3,100)
cdf = norm_rv.cdf(x)

plt.plot(x, cdf)
plt.ylabel("$F(X)$")
plt.xlabel("$x$")

# Отметим точку F(1)
plt.scatter([1], [norm_rv.cdf(1)], color="blue")
```

Можно найти вероятность $P(1 < X < 3) = F(3)-F(1)$

```python
norm_rv.cdf(3)-norm_rv.cdf(1)
```

```python
x = np.linspace(-4, 4, 100) # array диапазон
pdf = norm_rv.pdf(x)

plt.plot(x, pdf)
plt.ylabel("$f(x)$")
plt.xlabel("$x$")

# Нарисуем точки f(1), f(2)
plt.scatter([1, 3], [norm_rv.pdf(1), norm_rv.pdf(3)], color="blue")

# Добавим штриховку
xq = np.linspace(1, 3)
yq = norm_rv.pdf(xq)
plt.fill_between(xq, 0, yq, color="blue", alpha=0.2) # Кривые определяются по точкам ( x , y1 ) и ( x , y2 )
# Синия линия 
plt.axvline(1, color="blue", linestyle="--", lw=2)
plt.axvline(3, color="blue", linestyle="--", lw=2)
```

Можно найти квантиль(число q, такое что $P(X<=q)=\gamma$ ).
Найдем 5% и нарисуем его на картинке с плотностью.

```python
q = norm_rv.ppf(0.05)
# q - norm_rv.ppf(0.5) # Медиана
```

```python
x = np.linspace(-3, 3, 100) # array диапазон
pdf = norm_rv.pdf(x)

plt.plot(x, pdf)
plt.ylabel("$f(x)$")
plt.xlabel("$x$")

# Добавим штриховку
xq = np.linspace(-3, q)
yq = norm_rv.pdf(xq)
plt.fill_between(xq, 0, yq, color="blue", alpha=0.2)
# Синия линия 
plt.axvline(q, color="blue", linestyle="--", lw=2)
```

Из пакета stats можно использовать любое другое распределение. Например экспоненциальное распределение.

```python
exp_rv = stats.expon(scale=0.8)
x = np.linspace(0,4,100)
pdf = exp_rv.pdf(x)

plt.plot(x, pdf)
plt.ylabel("$f(x)$")
plt.xlabel("$x$")

xq = np.linspace(0.5, 2)
xy = exp_rv.pdf(xq)
plt.fill_between(xq, 0, xy, color = "blue", alpha = 0.5)
print(exp_rv.cdf(2)- exp_rv.cdf(0.5))
```

Функция экспоненциального распределения

```python
x = np.linspace(0, 4, 100)
cdf = exp_rv.cdf(x)

plt.plot(x, cdf)
plt.xlabel("$X$")
plt.ylabel("$F(X)$")
```

# Эмпиричесоке распределение

Работа с выборкой

```python
norm_rv = stats.norm(loc=0, scale=1) # Задали генератор
sample = norm_rv.rvs(1000) # генерация 1000 значений
```

```python
np.mean(sample) # Выборочное среднее
```

```python
np.var(sample) # Выборочная дисперсия
```

```python
np.std(sample) # Стандартное отклонение (корень из дисперсии)
```

```python
np.median(sample)
```

Построим гистограмму. Гистограмма - это эмпирическая оценка для плотности распределения

```python
plt.hist(sample, bins=50);
```

```python
x = np.linspace(-3, 3, 100)
pdf = norm_rv.pdf(x)

 # Плотность нормального распределения (lw = толщина)
plt.plot(x, pdf, lw=3)
# Гистограмма, параметр density отнормировал её
plt.hist(sample, bins=30, density=True)

plt.ylabel("$f(x)$")
plt.xlabel("$x$")
```

Построим по выборке эмпирическую функцию распределения

```python
# ECDF строит эмпирическую функцию по выборке
from statsmodels.distributions.empirical_distribution import ECDF

ecdf = ECDF(sample)

plt.step(ecdf.x, ecdf.y) # step - ступенчатая функция
plt.ylabel("$F(x)$")
plt.xlabel("x")
```

```python
x = np.linspace(-3, 3, 100)
# Теоретическая cdf
cdf = norm_rv.cdf(x)
plt.plot(x, cdf, label = "theoretical CDF")

# Эмпирическя cdf
ecdf = ECDF(sample)

plt.step(ecdf.x, ecdf.y, label = "empirical CDF") 
plt.ylabel("$F(x)$")
plt.xlabel("x")
plt.legend(loc="upper left")
```

# Генерация выборок

Будем генерировать значения с помощью np.random библиотеки numpy

```python
np.random.choice(np.arange(1,11),
                size=5,
                replace=False) # Выборка без повторений
```

Параметр replace базово равен True

```python
np.random.choice(np.arange(1,11),
                size=5,
                replace=True) # Выборка с повторениями
```

Можно расставить вероятность событий p и ииспользовать, например, массив из строк

```python
np.random.choice(["Карась","Плотва","Щука"],
                size=3,
                p=[0.5, 0.2, 0.3]) # p - вероятность
```

# Зёрна генерации (seed)

Выполняя команду stats.norm(loc=3, scale=2).rvs(5) на разных ПК получаем разные значения (по факту кста)

Чтобы получить одинаковые значения на двух различных ПК,, необходимо синхранизировать генераторы случайных чисел. Делается путем задания зерна (сида) генератора случайных чисел (любое целое число)

```python
# random_state - зерно генерации
stats.norm(loc=3,scale=2).rvs(5, random_state=42)
```
